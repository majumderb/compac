```python
import pickle
import random
import pandas as pd
```

```python
random.seed(2020)
```

```python
loc = 'gens/comet_hist_prior_all.pickle'

with open(loc, 'rb') as f:
    gens = pickle.load(f)
```

```python
gen_small = random.sample(gens, 100)
```

```python
persona = []
speaker_a1 = []
speaker_b1 = []
speaker_a2 = []
ours = []
gold = []

for g in gen_small:
    if len(g['history']) > 2:
        persona.append(g['persona'])
        speaker_a1.append(g['history'][-3])
        speaker_b1.append(g['history'][-2])
        speaker_a2.append(g['history'][-1])
        ours.append(g['generated'])
        gold.append(g['gold'])
```

```python
gen_df = pd.DataFrame(zip(persona, speaker_a1, speaker_b1, speaker_a2, ours, gold), columns=['p', 'c11', 'c12', 'c13', 'r1a', 'gold'])
```

```python
gen_df
```

```python
gen_df.to_csv('gens/sample_2.csv', index=False)
```

```python
# load gen 2

gen2 = pd.read_csv('gens/sample_3.csv', encoding='latin-1')
```

```python
gen2.head(1)
```

```python
compare = ['ours1', 'base', 'cbase', 'padial', 'gold']


c1_list = []
c2_list = []
c3_list = []
r1_list = []
r2_list = []
q1_list = []
q2_list = []
q3_list = []

for i, row in gen2.iterrows():
    for c in compare:
        c1_list.append(row['c11'])
        c2_list.append(row['c12'])
        c3_list.append(row['c13'])
        
        p = random.uniform(0, 1)
        if p > 0.5:
            r1_list.append(row['ours2'])
            r2_list.append(row[c])
            q1_list.append('q1_{}_{}'.format('our2', c))
            q2_list.append('q2_{}_{}'.format('our2', c))
            q3_list.append('q3_{}_{}'.format('our2', c))
        else:
            r1_list.append(row[c])
            r2_list.append(row['ours2'])
            q1_list.append('q1_{}_{}'.format(c, 'our2'))
            q2_list.append('q2_{}_{}'.format(c, 'our2'))
            q3_list.append('q3_{}_{}'.format(c, 'our2'))
```

```python
eval_df = pd.DataFrame(list(zip(c1_list, c2_list, c3_list, r1_list, r2_list, q1_list, q2_list, q3_list)), columns=['c11', 'c12', 'c13', 'r1a', 'r1b', 'q11', 'q12', 'q13'])
```

```python
eval_df.to_csv('gens/human_eval_2.csv', index=False)
```
